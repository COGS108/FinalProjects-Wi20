# Data Cleaning

### Pruning the Business Dataframe

Our data cleaning process has three main components: removing unnecessary information, moving around data to different columns to make them more accessible, and converting string data to numerical quantities. 

The first step is removing unnecessary columns, in particular identifying information for businesses. These columns include business names, aliases, Yelp ID's, phone numbers and URLs, since these do not contribute to our EDA in any meaningful way. 

Our second step is splitting up some data; in particular, we wanted to make separate latitude and longitude columns by extracting information from the coordinates column, where each entry was its own JSON object. This will be particularly helpful for us when we create geographic visualizations.

Lastly, we needed to convert some string data to numerical quantities. Yelp gives a restaurant's relative price on a scale from 1 to 3, where 1 is represented by a single dollar sign and 3 is represent by three dollar signs. We needed to transform these strings to make relative prices easier to analyze numerically.

```python
business_df.drop(columns=['alias', 'url', 'image_url', 'phone', 'display_phone', 'distance', 'name', 'location'], inplace=True)
business_df['stars'] = business_df['review_count'] * business_df['rating']
business_df['latitude'] = business_df['coordinates'].apply(lambda x: x['latitude'])
business_df['longitude'] = business_df['coordinates'].apply(lambda x: x['longitude'])
business_df.drop(columns=['coordinates'], inplace=True)
business_df['categories'] = business_df['categories'].apply(lambda cats: [cat['title'] for cat in cats])
business_df['price'] = business_df['price'].apply(lambda dollars : None if type(dollars) != str else len(dollars))
```

### Geodataframe Creation

For our geographic visualizations, we needed to transform our businesses dataframe into an identical geodataframe. To do this, we used the latitude and longtitude columns to produce a new coordinates column containing shapely.Point objects. These point objects can then be plotted on a latitude-longitude map, using a subsection of the San Diego region dataset as a background.

```python
#convert business_df into a geodataframe
business_df['coordinates'] = list(zip(business_df['longitude'], business_df['latitude']))
business_df['coordinates'] = business_df['coordinates'].apply(Point)
business_df = business_df.loc[(business_df['latitude'].isnull() == False) &
                                (business_df['longitude'].isnull() == False)]
business_gdf = gpd.GeoDataFrame(business_df, geometry = 'coordinates')
business_gdf.crs = san_diego_gdf.crs
```

### Average Number of Reviews for a Single Business per Drink

The bar plot below displays, as titled, the average number of reviews a business has per drink type. One clear benefit of performing this analysis is that it aids in equalizing the potential discrepancy that arises when a drink type has more reviews due to having more businesses than other drink types. When looking at the graph, it is evident that a business of the drink type ‘boba’ has the highest average number of reviews. However, before determining that boba is the most popular drink based on this graph alone, it is important to consider two other major factors. ### Distributions of Business Review Counts per Drink

For every drink type, we isolated the businesses corresponding to that drink. Then, we graphed the distribution of the number of reviews per business. The results were interesting. Across all drinks, the distribution of review counts per business was always unimodal and skewed right (positive skew). This was somewhat to be expected: most businesses on Yelp are not reviewed very often, with most having less than 1000 reviews. Only a select few, very popular businesses have extremely high review counts; these businesses are responsible for the extreme right skew. 

Interestingly enough, between the extremely popular businesses and the average businesses, there exist almost no businesses. This is especially prominent for alcohol. Our collective reasoning behind this is that for established drinks like alcohol, there are little “up and coming” businesses that are on a current trajectory to popularity. Most businesses have been around for a long time and either have widespread marketability or, in most cases, likely have a core base of customers restricted to the geographic region around them. In boba, however, the distribution is more varied: we see more businesses in between the leaders and the average. This could be because boba is a newer drink that has rapidly achieved popularity within the last decade, compared to alcohol and coffee which have been around much longer; thus, there are more boba businesses on the rise. Theoretically, as time passes, if boba continues to remain a mainstream drink, we should see the separation grow to look more like coffee and alcohol.

```python
drinks = business_df['drink'].unique()
drinks.sort()
rcvc = business_df.groupby('drink')['review_count'].value_counts()
rows = []
for i in range(len(rcvc)):
    key = rcvc.keys()[i]
    cnt = rcvc[i]
    for _ in range(cnt):
        rows.append([key[0], key[1]])
cdf = pd.DataFrame(np.array(rows), columns=['drink', 'review_count'])
cdf = cdf.astype({'review_count': np.int32})
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i, drink in enumerate(drinks):
    counts = cdf[cdf.drink == drink]['review_count']
    sns.distplot(counts, bins=20, ax=ax[i])
    ax[i].set(title='Average Review Count Distribution for {}'.format(drink))
```

### Distribution of Ratings per Drink 

The average rating of boba was 4.0 stars, coffee was 4.25 stars, and alcohol was 4.15 stars. This means coffee category Yelp reviews are on average more favorable than boba and alcohol Yelp reviews. Overall, the number of 4.0 vs 4.5 stars over all 3 categories were very close at ~670 4.0 stars and ~650 4.5 stars. From the large amount of 4.0 and above stars for all 3 categories where supposedly 2.5 - 3.0 stars should be the average, we wonder whether there is some sort of rating inflation either within the drink industries (based on this data alone) or whether raters/users/Yelpers (specifically) do not rate with 2.5 - 3.0 stars as the “middle” in mind. We do muse whether coffee shops have higher ratings due to actual drink quality or service quality. If that’s the case we wonder why alcohol does not have a 4.25 average as well, but from our experience boba shops have not have had as attentive of service and recommendations as coffee and alcohol.

```python
drinks = business_df['drink'].unique()
drinks.sort()
rrc = business_df.groupby('drink')['rating'].value_counts().sort_index()
rows = []
for i in range(len(rrc)):
    key = rrc.keys()[i]
    cnt = rrc[i]
    for _ in range(cnt):
        rows.append([key[0], key[1]])
rdf = pd.DataFrame(np.array(rows), columns=['drink', 'rating'])
rdf = rdf.astype({'rating': np.float32})
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i, drink in enumerate(drinks):
    counts = rdf[rdf.drink == drink]['rating']
    sns.distplot(counts, bins=10, ax=ax[i])
    ax[i].set(title='Average Rating Distribution for {}'.format(drink))
```

### Breakdown of Shared Categories Across All Drinks

These bar graphs display item categories that have been found boba, coffee, and alcohol places. This graph was built by looking at all the non-unique categories across drinks and then summing the total occurrences of each such category per drink. 

We can see that coffee businesses occupy the most of these categories, while boba and alcohol occupy substantially less, with alcohol businesses occupying the least. However, the vast presence of coffee in this graph can also be attributed to the sheer number of coffee shops and is also inflated by the ‘Coffee & Tea’ category. Since coffee businesses far outnumber the other businesses in these shared categories, it suggests that coffee is likely served in the widest variety of places. Another point of interest is that the majority of the most common categories in alcohol businesses were the weakest categories for the other drinks. This suggests that alcohol appeals to substantially different markets than the other two drinks. If we account for the difference in total businesses between coffee and boba, both coffee and boba have strong showings in the ‘Coffee & Tea’, ‘Bubble Tea’, and ‘Desserts’ categories. Meanwhile, alcohol displays strong showings in ‘American (New)’, ‘Breakfast & Brunch’, and ‘Sandwiches’, while still having a significant presence in ‘Asian Fusian’, ‘Specialty Food’, ‘Korean’, ‘Tapas/Small Plates’, and ‘Sushi Bars’.

```python
eecat = business_df.explode('categories')
unshared = eecat.groupby('categories')['drink'].nunique()
shared = []
for i in range(len(unshared)):
    if unshared[i] == 3:
        shared.append(unshared.keys()[i])
drinks = business_df['drink'].unique()
drinks.sort()
rows = []
for cat in shared:
    total = eecat[eecat.categories == cat]['categories'].count()
    cnts = eecat[eecat.categories == cat].groupby('drink')['categories'].count()
    for drink in drinks:
        rows.append([cat, drink, cnts[drink], total])
sdf = pd.DataFrame(rows, columns=['category', 'drink', 'businesses', 'total'])
sdf.sort_values(by='total', ascending=False, inplace=True)
sns.catplot('businesses', 'category', col='drink', data=sdf, kind='bar')
```

### Probability a Business Supports Delivery, Reservation, or Pickup

This bar plot displays the probability a certain drink business supports a type of transaction. We define (or rather Yelp defines) a transaction to either be delivery, reservation, or pickup. This graph was generated by aggregating the total number of businesses that allowed each transaction under each drink and then dividing by the total number of businesses for each drink. By removing the total number of businesses for each drink from the equation, we are able to see an unbiased view of how drink type directly affects transaction type.

We can see that delivery is the most likely transaction in all three drinks, followed by pick-up, and lastly reservations. Alcohol seems to be the least likely to offer any of the three transactions, while boba and coffee have an over 50% chance of offering delivery. This suggests that alcohol seems to primarily target walk-in customers. This would also be supported by the fact that alcohol businesses have a higher probability of offering in-person reservations. However, since there are significantly less boba shops then there are alcohol or coffee shops, it seems probable that boba shops would try to offer delivery transactions to reach a wider customer base. However, given the plentiful number of coffee businesses in the area and the probable chance that coffee shops offer delivery, it is likely that coffee shops might have the farthest reaching base. 

```python
drinks = business_df['drink'].unique()
drinks.sort()
rtdf = business_df.explode('transactions').dropna(subset=['transactions'])
rtdf['transaction'] = rtdf['transactions'].apply(lambda x : 'reservation' if x == 'restaurant_reservation' else x)
rtdf.drop(columns=['transactions'], inplace=True)
tdf = rtdf.groupby(['drink', 'transaction']).size().reset_index().rename(columns={0:'count'})
tdf['probability'] = tdf.apply(lambda x : x['count'] / len(business_df[business_df.drink == x['drink']]), axis=1)
ax = sns.barplot(x='drink', y='probability', hue='transaction', data=tdf)
ax.set_title('Probability a Drink Business Supports a Transaction Type')
```
