## Preperation for mapping

First, we prepare our data for mapping purposes. This setup requires slightly different management from other data analysis as we map data by ZIP codes instead of community. For mapping, we use the SANDAG data set which provides us with a list of reported crimes tagged with ZIP codes. Because the mapping data doesn't match 1:1 with our set of ZIP codes, careful cleaning must take place.

Most notably, we are forced to ignore reported crimes which aren't tagged with a ZIP code, or are tagged with a ZIP code outside of San Diego County. We deem this fair to do since such data is either erroneous or simply impossible to map.

```python
# Clean and prepare map data

## Create empty series to fill with postal codes of communities
map_df["AZip"] = ""
map_df["AZip"] = np.empty((len(map_df), 0)).tolist()

# Drop rows with NaN because it is not possible for us to plot the crime on the choropleth map
# if the zip code is missing. This action would not affect the outcome much in the choropleth map.
cdfd = cdf.dropna() # drop NA values (as explained in comment and explanation above)
cdfd['ZipCode']=cdfd['ZipCode'].astype(int) # convert floats to int (so ZIP codes don't end in .0)

# Determine what is missing between mapped ZipCodes and ZipCodes in the crime data
cSet=set(cdfd['ZipCode'].unique())
mSet=set(map_df['ZIP'].unique())
b=cSet-mSet
```

```python
# Manually add postal codes to map data
# These ZIP codes geographically map to these correct regions, but aren't previously in our imported mapping data
map_df.loc[map_df["ZIP"]=="91911"]["AZip"] = map_df.at[5,"AZip"].append(91909)
map_df.loc[map_df["ZIP"]=="91977"]["AZip"] = map_df.at[22,"AZip"].append(91976)
# 92012 doesnt exist
map_df.loc[map_df["ZIP"]=="92024"]["AZip"] = map_df.at[36,"AZip"].append(92023)
# Oceanside
map_df.loc[map_df["ZIP"]=="92054"]["AZip"] = map_df.at[46,"AZip"].append(92049)
map_df.loc[map_df["ZIP"]=="92054"]["AZip"] = map_df.at[46,"AZip"].append(92051)
map_df.loc[map_df["ZIP"]=="92054"]["AZip"] = map_df.at[46,"AZip"].append(92052)
# Vista
map_df.loc[map_df["ZIP"]=="92084"]["AZip"] = map_df.at[75,"AZip"].append(92085)
# Fallbrook
map_df.loc[map_df["ZIP"]=="92028"]["AZip"] = map_df.at[40,"AZip"].append(92088)
# 92100 doesnt exist
map_df.loc[map_df["ZIP"]=="92154"]["AZip"] = map_df.at[115,"AZip"].append(92158)
# 92371, 92543, 92592, 92924, 93244 and happened outside of ZIP code range for San Diego or don’t exist
```

```python
# Split data into violent and nonviolent
# These definitions for violent and nonviolent crime are set by the data provided in our ARGIS data set.
seriousCrime = cdfd[(cdfd['CM_LEGEND']=="HOMICIDE")|(cdfd['CM_LEGEND']=="SEX CRIMES")|(cdfd['CM_LEGEND']=="ASSAULT")|(cdfd['CM_LEGEND']=='WEAPONS')]
notSerious = cdfd[(cdfd['CM_LEGEND']!="HOMICIDE")&(cdfd['CM_LEGEND']!="SEX CRIMES")&(cdfd['CM_LEGEND']!="ASSAULT")&(cdfd['CM_LEGEND']!='WEAPONS')]

# Sum up all counts of violent crime
map_df["Count"]=0
c=seriousCrime["ZipCode"].value_counts().to_dict()
for i in c:
    if (i in map_df.ZIP.unique()):
        map_df.loc[map_df["ZIP"]==i,"Count"]+=c[i];
    else:
        for index,row in map_df.iterrows():
            if (i in map_df.loc[index,"AZip"]):
                map_df.loc[map_df["ZIP"]==i,"Count"]+=c[i]
```

```python
# get total count/ratio of violent crime
a=list(c.values())
total=np.sum(a)
map_df["LocalViolent/TotalViolent"]=map_df['Count']/total
 
# rename count -> violent
map_df.rename(columns={"Count":"Violent"},inplace = True)
```

```python
# Note: the following lines of code would be used to check to see if there are any
# unique zipcodes in the non-violent set, i.e. if any places had non-violent crime and no violent crime.
# There aren't any, so any possible addition of these are skipped.
# nZipSet=set(notSerious['ZipCode'].unique())
# c=nZipSet-mSet
# b
# 
# c

# Reset count to then be used for non-violent crime
map_df['Count']=0
 
# Sum up all counts of non-violent crime
d=notSerious["ZipCode"].value_counts().to_dict()
for i in d:
    if (i in map_df.ZIP.unique()):
        map_df.loc[map_df["ZIP"]==i,"Count"]+=d[i];
    else:
        for index,row in map_df.iterrows():
            if (i in map_df.loc[index,"AZip"]):
                map_df.loc[map_df["ZIP"]==i,"Count"]+=d[i]

# rename to be non-violent
map_df.rename(columns={"Count":"PropertyCrime"},inplace = True)
 
# setup a few more values to later be used by analysis
map_df["ViolentVsProperty"] = map_df["Violent"]/map_df["PropertyCrime"]
map_df["LocalProperty/TotalProperty"] = map_df["PropertyCrime"]/np.sum(list(d.values()))
 
# “fix” for divide by zero errors
map_df["ViolentVsProperty"].fillna(0,inplace=True)
```

Below is a method for cleaning up the ARJIS data tables to account for changing values. When a value is changed after entered, they would change the previously entered number to the negative value. Since we have no way of getting that information, we would just have to rely on the old data before it was changed. 

```python
def negative_cleanup(item):
    if isinstance(item, str):
        return item
    else:
        return abs(item)
```

The following method is a simple method to traverse the SANDAG data. The basic premise it it will go through every permutation of month and crime and find out how many crimes of that type occured in that month. It does this by filtering through of the available months, filter by how many times that crime occurs that month, and then sum them up into a simple dataset while also assigning those crimes as either violent or property crime. From there you take that dataset and then sum up the number of crimes marked as violent and number of times that crimes that are property.  

```python
def sandag_com(fd):
    violent_crimes = ['ASSAULT', 'SEX CRIMES', 'WEAPONS', 'HOMICIDE']
    index = ['DRUGS/ALCOHOL VIOLATIONS', 'THEFT/LARCENY', 'ASSAULT', 'VEHICLE BREAK-IN/THEFT', 'MOTOR VEHICLE THEFT', 'VANDALISM', 'BURGLARY', 'FRAUD', 'DUI', 'ROBBERY', 'SEX CRIMES', 'WEAPONS', 'ARSON', 'HOMICIDE']
    column=['AUG 2019', 'SEPT 2019', 'OCT 2019', 'NOV 2019', 'DEC 2019', 'JAN 2020']
    
    temp_summary = pd.DataFrame(index = index, columns=column)
    temp_180 = pd.DataFrame()
    fd['month'] = pd.DatetimeIndex(fd['Date']).month
 
    month = 8
    month_index = 0
    crime_index = 0
    while month != 2: 
        temp = fd[fd['month'] == month]
        temp_summary.at[index[crime_index], column[month_index]] = temp[temp['crime'] == index[crime_index]].shape[0]
        temp_summary.at[index[crime_index], 'Type'] = 'violent' if index[crime_index] in violent_crimes else 'property'
        crime_index = crime_index + 1
        if crime_index > 13:
            crime_index = 0
            month_index = month_index + 1
            if month == 12:
                month = 1
            else: 
                month= month + 1
    i = 0
    columns = temp_summary.columns
    while(i < len(columns) - 1):
        setemp = temp_summary
        temp_180.at['Violent Crime', columns[i]] = setemp.loc[setemp['Type'] == 'violent', columns[i]].sum()
        temp_180.at['Property Crime', columns[i]] = setemp.loc[setemp['Type'] == 'property', columns[i]].sum()
        i = i + 1
    
    return temp_180
```

This method is used to traverse a dataframe and grab a wanted amount of items to form a smaller dataframe based on the inputs given. For example, you would pass in the unformated years for a a dataframe from say Carlsbad, pass in the formated list, and the datframe you want. It will grab all entries matching the unformatted years in Carlsbad in the list, and make a new dataframe based on the formatted years. Since it formats it in such a way, it will also change it so that it will only have add up the crimes in violent and property and make that an entry. 

```python
def comm_monthly(yearn, yearf, arjis):
    temp = pd.DataFrame(columns = yearf, index = ['Violent Crime', 'Property Crime'])
    
    i = 0
    for i in range(len(yearn)):
        temp.at['Violent Crime', yearf[i]] = arjis.loc[arjis['Type'] == 'violent', yearn[i]].sum()
        temp.at['Property Crime', yearf[i]] = arjis.loc[arjis['Type'] == 'property', yearn[i]].sum()
        
    return temp
```

This method is used to set some values from a list to a dataframe. You give a dataframe, the city/community you want to change along with what crime you want, and assign it something from a list of numbers. 

```python
def set_num(dataframe, city, column, numbers):
    for i in range(len(city)):
        dataframe.at[city[i], column] = numbers[i]
    return dataframe
```

```python
rows = ['Carlsbad', 'Chula Vista', 'Coronado', 'El Cajon', 'Escondido', 'La Mesa', 'National City', 'Oceanside','San Diego', 'Sheriff']
want = ['Property Total', 'Violent Total']
pval = ['Property P', 'Violent P']

comp = pd.DataFrame()
want = ['Property Total', 'Violent Total']
pval = ['Property P', 'Violent P']
for i in rows:
    whatever, comp.at[i, pval[0]] = x.proportions_ztest(np.array([sandag_data.at[i, want[0]], sandag_data.at[i, 'Total']]), np.array([arjis_data.at[i, want[0]],arjis_data.at[i, 'Total']]))
    whatever, comp.at[i, pval[1]] = x.proportions_ztest(np.array([sandag_data.at[i, want[1]], sandag_data.at[i, 'Total']]), np.array([arjis_data.at[i, want[1]],arjis_data.at[i, 'Total']]))
    comp.at[i, 'Violent Sig'] = 'sig' if comp.at[i, pval[1]] < 0.05 else 'not sig'
    comp.at[i, 'Property Sig'] = 'sig' if comp.at[i, pval[0]] < 0.05 else 'not sig'
    
comp
```

We followed a null hypothesis that the two datasets should not be that different when it comes to crime, and an alternate hypothesis that they would not be different. In order to calculate this, we had to do a statisical test with respect to proportionality. In this case, a significant difference would be a p-val of less than 0.05. Overall, we see that the crime rates are not that significantly different, but a couple of times it does significantly different. This could be due to factors such as the incompatiability of the crime definitions, along with the possible changes in violent crime.
