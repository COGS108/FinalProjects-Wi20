MCAT scores range from 472 to 528, 472 being the lowest score possible and 528 being a perfect score. Since 2017 the average for the MCAT scores have increased. The trend shows that since 2017 the average for the MCAT scores have increased, which means that students have been performing better each year. Not only does the admissions department look at gpa and MCAT scores, however they also take into consideration extracurricular activities, clinical experience, and shadowing a physician into acceptances. They want to see what the student is doing for the community, and the activities done to show why they are so passionate in becoming a physician. 
 
In researching gpa/MCAT scores in the U.S. Medical Schools we found that there were differences in acceptances throughout races/ethnicities. The data provided by the AAMC.org showed that other races/ethnicities had lower gpa’s and MCAT scores compared to Asians or White Americans. # Data Cleaning

STEPS: 
Many tables we scraped from websites were formatted very strangely, so part of our cleaning process was to format our tables such that each row and column matched up with their index

1. Convert csv, pdf, and webpages into pandas dataframes
2. Name/rename columns so that their rows and columns match up
3. Remove rows and columns with data that is not relevant to our questions (i.e. applicant #, matriculant # and %, etc., we only considered acceptance rates with the exception with table A-8 where analyzing the number of male and female applicants by race was necessary)
4. Combine rows with similar racial data (i.e. if we had rows like "Asian Only", "Asian, Black or African American", and "Asian, White" we combined them into an "Asian" row)
5. Strip non numerical characters from dataframes
6. Convert dataframe values to floats
7. Set indexes to race/ethnicity or by GPA/MCAT Scores if race/ethnicity was not a present column in the dataframe

Table accvrace cleaning

```python
# rates of acceptance and matriculation to U.S. med schools by race/ethnicity, 2019-2020
accvrace_df = accvrace_table_list[0]

# drop irrelecant columns and rows
accvrace_df = accvrace_df.drop([0, 1])
accvrace_df = accvrace_df.drop([6], axis=1)
accvrace_df.columns = [
                              'Race/Ethnicity', 
                              'Applicants', 
                              'Acceptees', 
                              '% Accepted',
                              'Matriculants',
                              '% Matriculated'
                              ]
                              
# rename rows and strip non number characters
for row in accvrace_df.iterrows():

  # rename rows to have similar names for grouping later
  if (row[1]['Race/Ethnicity'].startswith('American Indian')):
    row[1]['Race/Ethnicity'] = 'American Indian'
  elif (row[1]['Race/Ethnicity'].startswith('Asian')):
    row[1]['Race/Ethnicity'] = 'Asian'
  elif (row[1]['Race/Ethnicity'].startswith('Black')):
    row[1]['Race/Ethnicity'] = 'Black/African American'
  elif (row[1]['Race/Ethnicity'].startswith('Hispanic')):
    row[1]['Race/Ethnicity'] = 'Hispanic/Latinx'
  elif (row[1]['Race/Ethnicity'].startswith('White')):
    row[1]['Race/Ethnicity'] = 'White'

    # strip non number characters
  row[1]['% Accepted'] = row[1]['% Accepted'].replace('%', '')
  row[1]['% Matriculated'] = row[1]['% Matriculated'].replace('%', '')
  
accvrace_df = accvrace_df.set_index(['Race/Ethnicity'])
accvrace_df = accvrace_df.astype(int)
```

Table: Acceptance vs. In/Out of state residence cleaning

```python
# create df from table list
accvresidence_df = accvresidence_table_list[0]

# format df and remove any unessessary info
accvresidence_df.columns = accvresidence_df.iloc[0]
accvresidence_df = accvresidence_df.drop(0)
accvresidence_df = accvresidence_df.drop(['In-State Advantage', 'State'], axis=1)
accvresidence_df = accvresidence_df.set_index('Medical School')


# strip all non numerical characters
for row in accvresidence_df.iterrows():
    row[1]['Overall Acceptance Rate'] = row[1]['Overall Acceptance Rate'].replace('%', '')
    row[1]['In-State Acceptance Rate'] = row[1]['In-State Acceptance Rate'].replace('%', '')
    row[1]['Out of state Acceptance Rate'] = row[1]['Out of state Acceptance Rate'].replace('%', '')

# convert to all float types
accvresidence_df = accvresidence_df.astype(float)

# out of state acceptance rate list
oos_acc = []
# in state acceptance rate list
is_acc = []

# length of table
numrows = len(accvresidence_df)

# find average total in and out of state acceptance rates
for row in accvresidence_df.iterrows():
    is_acc.append(row[1]['In-State Acceptance Rate'])
    oos_acc.append(row[1]['Out of state Acceptance Rate'])

# calculate averages
oos_sum = np.sum(oos_acc)
is_sum = np.sum(is_acc)
oos_avg = float(str(round(oos_sum / numrows, 2)))
is_avg = float(str(round(is_sum / numrows, 2)))

residence_avg = [is_avg, oos_avg]
```

cleaning table A-8

```python
# build dataframe from csv
sexvrace_df = pd.read_csv(sexvrace_path_out, header=None)

# drop irrelevant rows
sexvrace_df = sexvrace_df.drop([0, 1])

# rename columns
sexvrace_df.columns = [
    'Race/Ethnicity',
    '2016-2017 Men',
    '2016-2017 Women',
    '2016-2017 Total',
    '2017-2018 Men',
    '2017-2018 Women',
    '2017-2018 Total',
    '2018-2019 Men',
    '2018-2019 Women',
    '2018-2019 Total',
    '2019-2020 Men',
    '2019-2020 Women',
    '2019-2020 Total'
]

# extract only data from 2019-2020's applicant pool
sexvrace_df = sexvrace_df[[
    'Race/Ethnicity',
    '2019-2020 Men',
    '2019-2020 Women',
    '2019-2020 Total',
]]


# rename rows and strip non number characters
for row in sexvrace_df.iterrows():

  # rename rows to have similar names for grouping later
  if (row[1]['Race/Ethnicity'].startswith('American Indian')):
    row[1]['Race/Ethnicity'] = 'American Indian'
  elif (row[1]['Race/Ethnicity'].startswith('Asian')):
    row[1]['Race/Ethnicity'] = 'Asian'
  elif (row[1]['Race/Ethnicity'].startswith('Black')):
    row[1]['Race/Ethnicity'] = 'Black/African American'
  elif (row[1]['Race/Ethnicity'].startswith('Hispanic')):
    row[1]['Race/Ethnicity'] = 'Hispanic/Latinx'
  elif (row[1]['Race/Ethnicity'].startswith('White')):
    row[1]['Race/Ethnicity'] = 'White'

  # strip non number characters
  row[1]['2019-2020 Men'] = row[1]['2019-2020 Men'].replace(',', '')
  row[1]['2019-2020 Women'] = row[1]['2019-2020 Women'].replace(',', '')
  row[1]['2019-2020 Total'] = row[1]['2019-2020 Total'].replace(',', '')
  
# set the index to be race/ethnicity
sexvrace_df = sexvrace_df.set_index('Race/Ethnicity')
sexvrace_df = sexvrace_df.astype(int)

# raname columns because we don't need to specify the year anymore
sexvrace_df.columns = [
    'Men',
    'Women',
    'Total'
]

# combine similar races together 
sexvrace_df = sexvrace_df.groupby(sexvrace_df.index, sort=False).sum()
```

Cleaning table A-18

```python
race_scores_df = pd.read_csv(race_scores_path_out, header=None)
race_scores_df = race_scores_df.drop(columns=[0, 1])
race_scores_df.columns=[
                        'American Indian',
                        'Asian',
                        'Black/African American',
                        'Hispanic/Latinx',
                        'Native Hawaiian or Other Pacific Islander',
                        'White',
                        'Other',
                        'Multiple Race/Ethnicity',
                        'Unknown Race/Ethnicity',
                        'Non-U.S. Citizen and Non-Permanent Resident',
                        'Total'
]

race_scores_df = race_scores_df.transpose()
race_scores_df = race_scores_df[[8, 9, 10, 11, 14, 15]]
race_scores_df.columns = [
                          'AvgMCAT',
                          'MCATsd',
                          'AvgScienceGPA',
                          'ScienceGPAsd',
                          'AvgTotalGPA',
                          'TotalGPAsd'
]
race_scores_df.index.name = 'Race/Ethnicity'
```

Clean table A-23

```python
# build df from csv
gpavmcat_df = pd.read_csv(gpavmcat_path_out, header=None)

# rename columns
gpavmcat_df.columns = [
    'GPA',
    'Groups',
    '484', 
    '488', 
    '492', 
    '496', 
    '500',
    '504',
    '508',
    '512',
    '516',
    '520',
    'AvgAccRate'
]

# holds the gpa that each acceptance rate row will represent
gpa = 0

# drop rows that don't have acceptance rate
for row in gpavmcat_df.iterrows():
    if row[1]['Groups'] == "Acceptees":
        gpa = row[1]['GPA']
        gpavmcat_df = gpavmcat_df.drop(axis=0, index=int(row[0]))
         
    if row[1]['Groups'] == "Applicants":
        gpavmcat_df = gpavmcat_df.drop(axis=0, index=int(row[0]))
        
    # fill gpa value into the acceptance rate row
    if row[1]['Groups'] == "Acceptance rate %":
        gpavmcat_df.at[row[0], 'GPA'] = gpa
   
# create new GPA labels so they can be worked on later
gpa_vals = [
            '3.9',
            '3.7',
            '3.5',
            '3.3',
            '3.1',
            '2.9',
            '2.7',
            '2.5',
            '2.3',
            '2.1',
            '1.9',
            'AvgAccRate'
]

gpavmcat_df['GPA'] = gpa_vals



# replace '-' and na values with 0
gpavmcat_df = gpavmcat_df.replace(to_replace='-', value='0')
gpavmcat_df = gpavmcat_df.fillna(0.0)

# drop irrelevant 'Groups' column
gpavmcat_df = gpavmcat_df.drop(columns=['Groups'])

# set index as the gpa column
gpavmcat_df = gpavmcat_df.set_index(['GPA'])

# convert all vals to floats
gpavmcat_df = gpavmcat_df.astype(float)

# revert index
gpavmcat_df = gpavmcat_df.reset_index()
```

use GPA as a predictor for acceptance rate based on race

```python
# get the relationship between just gpa and acceptance rate to see if the average gpas of different races correctly predict acceptance rates
gpavapp_df = gpavapp_df.astype(float)
outcome, predictors = patsy.dmatrices('AvgAccRate ~ GPA', gpavapp_df)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

Plot Linear and Polynomial Regression model

```python

# Plot the orginal data (as before)
sns.scatterplot(x='GPA', y='AvgAccRate', data=gpavapp_df, s=50)

# Generate and plot the model best fit line
xs = np.arange(gpavapp_df['GPA'].min(), gpavapp_df['GPA'].max() + .1)
ys = 28.3636 * xs + -62.8273
plt.plot(xs, ys, '--k', linewidth=2, label='Linear Model')

# Generate and plot model best fit polynomial
p = np.polyfit(
    x=gpavapp_df['GPA'], 
    y=gpavapp_df['AvgAccRate'], 
    deg=3
    )

plt.plot(
    gpavapp_df['GPA'], 
    np.polyval(p, gpavapp_df['GPA']), 
    'b--', label='Polynomial Model'
    )

# add labels 
plt.xlabel('Average GPA')
plt.ylabel('Average Acceptance Rate (%)')
plt.legend()
```

use to model to predict acceptance rates from the GPA scores of different races and plot the predictions

```python
predicted_rates = []
pz = np.poly1d(p)

# predict acceptance rates based on the model
for row in race_scores_df.iterrows():
    gpa = row[1]['AvgTotalGPA']
    gpa = float(gpa)
    predicted_rates.append(float(str(round(pz(gpa), 2))))

      
# build barplot to visualize acceptance rates by race categorized
accvrace = sns.barplot(
    x=race_scores_df.index,
    y=predicted_rates
)

# rotate x axis labels so we can read them
accvrace.set_xticklabels(
    labels=accvrace.get_xticklabels(), 
    rotation=90
)

# set title of barplot
accvrace.set(
    title="Acceptance Rate Predicted by GPA",
    ylim=(20, 45),
    ylabel="Acceptance Rate (%)"
)

plt.show()
```

Compare actual acceptance rates and predicted acceptance rates according to the GPA of different races

# Conclusion & Discussion

The question we sought to answer was whether US medical school applicants with similar MCAT/GPA scores will vary significantly in matriculation rates based on the applicant’s race. The inspiration for choosing this topic came from pre medical students within our group that have been posed with this question before and found no clear answer. Through the collection and analysis of the appropriate data, we sought to objectively answer this question and to shed light on the significance of the effects that race has on matriculation rate. We collected data pertaining to medical school applicants' race, MCAT, GPA, in-state applicants, out of state applicants, US applicants, non US applicants, gender, and associated matriculation rates. This data was collected through data scraping of the various websites listed in the above dataset section and the data underwent manipulation to fit the specifications for tidy data which improved its accessibility for further analysis. The analysis of the data focused on revealing differences between observed matriculation rates for specific races and predicted matriculation rates based upon MCAT/GPA scores. The observed matriculation rates across different races was shown to be relatively uniform with only slightly less African American and Native Hawaiian matriculants; the observed matriculation rates are as follows: American Indian(52%), Asian(44%), African American(38%), Hispanic(42%), Native Hawaiian(31%), White(44%), and non-citizen(17%). We concluded that non Us citizens, regardless of their race matriculated at significantly lower rates. We then used the collected MCAT/GPA scores and the corresponding matriculation rates,regardless of race, to predict the expected acceptance rates. We predicted that the different races would matriculate as follows: American Indian(26%), Asian(41%), African American(23%), Hispanic(29%), Native Hawaiian(32%), White(43%), and non-citizen(42%). By taking the difference between observed and predicted acceptance rates we can observe percent increases and decreases from the predicted values as follows: American Indian(+26%), Asian(+3%), African American(+15%), Hispanic(+13%), Native Hawaiian(-1%), White(-1%), and non US citizen(-25%). From this we can see that the observed and predicted acceptance rates across the specified races differed significantly among American Indians, African Americans, and Hispanics, while the matriculation rates for Whites, Asians, and Native Hawaiians showed nonsignificant differences.  Although Native Hawaiians and Non U.S. citizens would seem like minorities, we can attribute their lower actual acceptance rate to applying to schools out of state and out of country which is known to decrease acceptance rates. 

Some potential limitations that may affect the validity of our results could be our inability to control for the many variables that are considered when medical school boards choose their future students. According to Justine Falcone, an author from Med School Tutors, the top five most important factors considered in medical school admissions are Academic Coursework and Performance, Medical College Admissions Test (MCAT), Extracurricular Activities/Volunteer Experience, Letters of Recommendation, and Interview Skills (Med School Tutors, 2019). This is not a comprehensive list but it gives us an idea of how complex the medical school admissions decision making process is. To answer our proposed question we collected data pertaining to applicants with similar MCAT/GPA scores across several different races to see for example, if African American applicants with a 510 MCAT and 3.7 GPA matriculate at significantly different rates than Asian Americans with similar MCAT and GPA scores. Here we are controlling for MCAT and GPA scores by keeping them constant while comparing these scores across many races to see if there are significant changes in the matriculation rate. The problem with this is that we are only controlling for the MCAT/GPA scores while allowing the numerous other variables to affect our measurement of the differences in acceptance rates. This leads us to the question of whether we are measuring what we think we are measuring. For future experiments on this subject, the necessary variables will need to be controlled for to increase the validity of the results. The results of this experiment will allow medical school applicants to assess their competitiveness compared to others with similar MCAT/GPA scores on the basis of race. Accurate Knowledge of one's competitiveness compared to others is valuable to applicants when deciding which medical schools to apply to.
