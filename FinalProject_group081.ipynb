## Background and Prior Work

You can't teach an old dog new tricks is a popular idiom exemplifying the fact that it is more difficult to teach an older individual something new, especially because they are used to their own mindset, habits, and traits for so long. It’s a widely-used proverb that has lasted around five hundred years, with its earliest citation in print in The boke of husbandry (1534) by John Fitzherbert (1). Although this saying has prevailed for about five centuries, is it really true? Do old people learn slower than younger people?

On the other hand, there is a psychological idea that childrens’ brains are like sponges, soaking up a vast amount of information from their surroundings in a short amount of time. According to Abdul Malik Muftau at the Teachers College graduate school at Columbia University, children in youth sport teams tend to learn very fast compared to adults who are learning the same thing in their respective sport (2). It’s interesting to see how this could be possible. In “early stages of neuronal development, the brain exhibits more plasticity, and as a result children seem to absorb everything around them (language being one of the most notable examples)” (3). More specifically, by age two or three, infants enter a period called “Exuberant Synaptogenesis, where they have more than twice as many synaptic connections than adults” (3). Because “long-term memories are stored in the brain in the form of synapses”, memory formation happens when these connections are formed or strengthened (4). However, the “brain actually goes back and reduces the number of connections after [the Exuberant Synaptogenesis] because a functioning adult, set in [his/her/their] ways, doesn't need as many” (3). Although psychology and biology shows how children form memories more effectively and rapidly, does it matter whether one learns something earlier or later in life? More importantly to us, does this matter if you want to become a successful software engineer?

With the rise of technology and its company counterparts, the potential occupation as an engineer has increasingly become appealing in recent years. According to the U.S. Bureau of Labor Statistics, “employment of software developers is projected to grow twenty-one percent from 2018 to 2028, which is faster than the average for all occupations” (5). Many prospective college and university students see Computer Science and its equivalents as a fast and reliable route to gaining a stable job and income. As the competition in technology and the demand of software engineers increases, tech companies have accommodated accordingly by providing extra benefits, such as “relocation assistance, tuition reimbursement, wellness stipends” (6) to even micro kitchens to attract the “best and brightest” employees and maintain a “strong employer brand” (7). The stable income, the huge job market, and the emphasis on work culture and employee satisfaction have driven people from different backgrounds and ages to study towards becoming a software engineer, including the current youth.

Many companies such as iD Tech and FunTech are using this logic (the job market and benefit factor) to encourage parents to sign their children up for programming related courses and camps that many public schools do not offer in their curriculum. Even huge tech corporations such as Google host and encourage coding programs for younger audiences, ranging from elementary to high school students. Although many parents buy into these camps and programs, how does coding at an earlier age really affect the future of these potential software engineers?

Not many studies have been done on this topic. A notable study that relates the most with our question is research done by Amanda Strawhacker and Marina Umaschi Bers with the Association for Educational Communications and Technology. In 2018, they “investigated cognitive domains and computer programming knowledge in young children” to see what children learn when they begin learning how to code (8). In their research, they created a six-week learning intercession in which children engaged with Scratch Jr., a programming environment made for younger audiences, on an iPad, and then tested them on their programming knowledge with a numerical scored response form. Through their method, Strawhacker and Bers concluded that “children in [their] study demonstrated a range of mastery and creativity through coding, but none showed a complete lack of understanding about the most foundational programming concepts after the intervention” (8). Although it is not known from this research whether coding at an earlier age affects the student in the long-term scope of career success and happiness, it does support the fact that coding for children helps cognitive development in general.

Considering this developmental benefit, the extreme push towards teaching people how to code at a younger age and the mass assumption that learning earlier is more beneficial than learning later in life, we hope to realize through this data science project whether or not learning programming earlier really has an edge, not only in knowledge and industry, but in life and their happiness.

## Data Cleaning

### Dropping unnecessary columns 

```python
# List all columns in the data frame
var_19 = list(data_2019.columns)

# Drop columns that is not relevant to our hypothesis
df_19 = data_2019.drop(columns=['Respondent', 'MainBranch', 'Employment', 'Country', 
                                'EdLevel', 'EduOther', 'OrgSize', 'DevType', 
                                'CurrencySymbol', 'CurrencyDesc', 'CompTotal', 
                                'CompFreq', 'WorkWeekHrs', 'LanguageDesireNextYear', 
                                'DevEnviron', 'OpSys', 'Gender', 'Trans', 'Sexuality', 
                                'Ethnicity'])

# Show the first 5 index of our updated dataframe
df_19.head()
```

### Dealing with Na and Null Values

```python
df = df_19.dropna()
df.head()
```

### Standardize Values in our Columns

```python
# checking if our column has any discrepancy
# It does, so will apply standardizing function
print(df['YearsCode'].unique())

# Separator for easy visualization
print('---------------------------------------------------------------------')

# This column has some discrepancy as well
print(df['YearsCodePro'].unique())

#Creating a standardizing function
def standardize_Years(string):
    
    if 'Less' in string:
        out = '1'
    elif 'More' in string:
        out = '50'
    else:
        out = string
        
    return out

# Apply function two both columns to standardize them
df['YearsCode'] = df['YearsCode'].apply(standardize_Years)
df['YearsCodePro'] = df['YearsCodePro'].apply(standardize_Years)

# Find unique values in this column as well
df['Age1stCode'].unique()

# Creating a standardizing function
def standardize_Age1stCode(string):
    
    if 'Younger' in string:
        out = '5'
    elif 'Older' in string:
        out = '85'
    else:
        out = string
        
    return out

# Apply function to column data
df['Age1stCode'] = df['Age1stCode'].apply(standardize_Age1stCode)
```

### Fixing numbers in columns

```python
# Rounds values in column
df = df.round()

# Prints unique values to see if they are rounded
df['Age'].unique()
```

### Dealing with Categorical Data

```python
# Function to convert catagorical data to integers to find means
def convertInt(string):
    out = string
    if string == 'Very dissatisfied':
        out = 1
    elif string == 'Slightly dissatisfied':
        out = 2
    elif string == 'Neither satisfied nor dissatisfied':
        out = 3
    elif string == 'Slightly satisfied':
        out = 4
    elif string == 'Very satisfied':
        out = 5
    return out

# Create new columns of converted catagorical data to ints
df['CareerSatInt'] = df['CareerSat'].apply(convertInt)
df['JobSatInt'] = df['JobSat'].apply(convertInt)
```

```python
df.JobSat = pd.Categorical(df.JobSat, categories=['Very dissatisfied', 'Slightly dissatisfied', 'Neither satisfied nor dissatisfied',
                                                     'Slightly satisfied', 'Very satisfied'], ordered=True)
df.CareerSat = pd.Categorical(df.CareerSat, categories=['Very dissatisfied', 'Slightly dissatisfied', 'Neither satisfied nor dissatisfied',
                                                           'Slightly satisfied', 'Very satisfied'], ordered=True)
df.sort_values(['CareerSat', 'JobSat'], inplace=True)

df.CareerSat = pd.Categorical(df.CareerSat, categories=['Very dissatisfied', 'Slightly dissatisfied', 'Neither satisfied nor dissatisfied',
                                                           'Slightly satisfied', 'Very satisfied'], ordered=True)
```

### Standardize Values in our Columns Part 2

```python
# Creating a standardizing function
def standardize_LanguageWorkedWith(string):
    
    return string.count(';') + 1

# Apply function two both columns to standardize them
df['LanguageWorkedWith'] = df['LanguageWorkedWith'].apply(standardize_LanguageWorkedWith)
```

### Converting String Types into Integers

```python
df['YearsCode']    = pd.to_numeric(df['YearsCode'])
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'])
df['Age1stCode']   = pd.to_numeric(df['Age1stCode'])
```

### Renaming Columns into Readable and Understandable Columns

```python
# Rename columns
df = df.rename(columns={'LanguageWorkedWith': 'NumLanguageWorkedWith'})

# Resting index values to have it more organized
df = df.reset_index(drop=True)
```
