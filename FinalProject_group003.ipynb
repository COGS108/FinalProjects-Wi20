# Data Cleaning

Step 1: Create four useful columns for our study
- Column 1 (Accomodates):because we are interested in the price per person, we need to include the accommodates number per house
- Column 2-4 (price, neighbourhood_cleaned, amentities): We only need those three factors because we are only interested in the relationship between price and ratings, the relationship between amentities and ratings, and the relationship between specific neighborhood and ratings.
- Column 5 (ratings): The dependent variable of out project
- Column 6-7(bedrooms, bathrooms): is the ratio of bedroom and bath room we want to find

```python
# choose columns we need to use.
all_data = all_data[['accommodates', 'price', 'neighbourhood_cleansed', 'amenities', 'review_scores_rating','bedrooms','bathrooms']]
all_data
```

```python
# test if there are any null values 
all_data.isnull().sum()
```

* Step 2: Drop all the N/A values
* Step 3: Clean the price column: drop all dollar signs in the dataframe
* Step 4: Count the number of amentities
* Step 5: Extract outliers 
    

```python
# it is, since the review_scores_rating is the expected value we are looking at, the null values are meaningless to us.
# Therefore, we need to drop Na values
all_data = all_data.dropna()
all_data

#First, clean the price column and change the type of price column to np.float64, which can help us to calculate the price per person.

# (1) we build a help function which can drop '$' sign and change the type. 
def cleaning_function (str):
    ans = str[1:].replace(',', '')
    return ans

# (2) apply function to the price column in the all_data dataframe.
all_data['price'] = all_data['price'].apply(cleaning_function).astype(np.float64)

#second we will add a new column named price per person, which is the important factor we will observe. 
all_data['price_per_person'] = all_data['price']/all_data['accommodates']

#third, since the amenities column contains a string which has all amenities in each rows, we need to count how many amenities in each string 
#and add a new column named number_of_amenities, which help us to find the relationship between number_of_amenities and the ratings.

#(1) we build a helper function which can help us count the number of amenities in a string. 
def number_of_amenities(str):
    num = str.count(',')+ 1
    return num
#(2) then we apply the function to the amenities column and add the result column named number_of_amenities in the all_data dataframe.
all_data['number_of_amenities'] = all_data['amenities'].apply(number_of_amenities)

#(3) convert the ratio
all_data = all_data[all_data['bathrooms']!=0]
all_data['ratio'] = all_data['bedrooms']/all_data['bathrooms']

#then, keep the columns we need and drop columns we donot need anymore.
all_data = all_data[['neighbourhood_cleansed', 'price_per_person', 'number_of_amenities', 'review_scores_rating','ratio']]

#Besides, when we observe the relationship between price_per_person and review_scores_rating, we find that there are some outliers
#in the price_per_person column, we try to use np.percentile to solve this problem. We try to keep 1%- 99% values in the column to avoid
#outliers problem.
#we want to find the middle class affordable price airbnb.
minimum = np.percentile(all_data['price_per_person'], 1)
maximun = np.percentile(all_data['price_per_person'], 99)
all_data = all_data.loc[(all_data['price_per_person'] >= minimum) & (all_data['price_per_person'] <= maximun)]
```

```python
# first, we will consider about the relationship between price per person and review_scores_rating. 
price = pd.DataFrame(all_data.groupby('price_per_person').mean()['review_scores_rating'])
price['price_per_person'] = price.index
g = sns.jointplot(x="price_per_person", y="review_scores_rating", data=price, kind='reg', joint_kws={'line_kws':{'color':'red'}})
g
```

```python
# from the plot above, we find that the regression line in the graph only have a tiny postive climbing tendency, so we decide to use the linear
#regression model to test the score of the relationship. 

x = np.array(price['price_per_person']).reshape((-1,1))
y = np.array(price['review_scores_rating'])
lm = LinearRegression()
model = lm.fit(x,y)
lm.score(x,y)

#we find that the score is too low, which means that there is almost no relationship between price_per_person and review_scores_rating.
```

```python
#I want to find the ratio of the bedrooms with bathrooms to see is it related to the review scores rating.
ratio = pd.DataFrame(all_data.groupby('ratio').mean()['review_scores_rating'])
ratio = ratio.reset_index()
g = sns.jointplot(x="ratio", y="review_scores_rating", data=ratio, kind='reg', joint_kws={'line_kws':{'color':'red'}})
g
```

```python
x1 = np.array(ratio['ratio']).reshape((-1,1))
y1 = np.array(ratio['review_scores_rating'])
graph1 = sns.regplot(x=x1, y=y1)
lm = LinearRegression()
model = lm.fit(x1,y1)
lm.score(x1,y1)
#Since 0.004 is too low, we can say that there is almost no relationship between 
#the ratio of the bedrooms with the bathrooms and the review scores rating.<br>

#Then, let us check the relationship between number of amenities and review_scores_rating. 
#Since we have a huge dataset and there are only at most 104 numbers of amenities in the Airbnb house in the dataset, 
#we try to use the groupby to find the mean of review_scores_rating in each number of amenities.
```

```python
#Then, let us check the relationship between number of amenities and review_scores_rating. Since we have a huge dataset and there are only
#at most 104 numbers of amenities in the Airbnb house in the dataset, we try to use the groupby to find the mean of review_scores_rating in
#each number of amenities.

#first, we use the groupby to find the mean of review_scores_rating in each number of amenities.
amenities = pd.DataFrame(all_data.groupby('number_of_amenities').mean()['review_scores_rating'])
#We should add the column "number_of_amenities" in the amenities dataframe
amenities['number_of_amenities'] = amenities.index
amenities
```
