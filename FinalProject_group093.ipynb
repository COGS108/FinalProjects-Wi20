# Research Question

In a period when opportunities for better social equality are being offered more and more in the United States, we are interested in asking the following question: Are the recent admission rates of women in STEM careers in the United States fair?

## Background and Prior Work

The question is interesting because as our society gets more developed each and every day, we often wonder: “Are we all receiving the same equality in our school and workplace?” To answer the question, we decided to research the admission rate of women in STEM careers in the US.

STEM careers consist of careers in Science, Technology, Engineering, Math and Medicine. Basically, they make up a high demand globally due to a rise in living standards and technologies (Reference 1). While there is an increasing number of women in fields like medicine and law, women in STEM major seem to have a more difficult time choosing to be in this field (Reference 2). One of the reasons for that is because of the following barriers: gender bias and race stereotypes (Reference 3). For example, people used to think that only men are good at math and science, this plays a role in creating an emotional barrier for young women who are interested in the field (Reference 6). According to the article “8 Statistics and Facts about Women in STEM,” Silva states, “The percentage of women with STEM degrees has dropped from 25% to 24%. If the trend continues, this number is going to be even lower in the future." (Reference 5). The author argues that the number of men choosing STEM major surpasses the number of women who does and if this continues, women will continue to be unpresented in the STEM field.

Women and men do not receive the same equality when it comes to awarded grants. In a study done by US National Institute of General Medical Sciences, it was concluded that only one third of the women researchers received the grants while half of the participants were women (Reference 4). In the article "Women are less visible in STEM: Why?" it states that the women who are interested in STEM fields tend to “leak” or change their careers away from STEM as they become older due to variety of factors from social stereotypes to hiring discrimination (Reference 4). In addition, ethnicity is also one of the important factors that can alter the number of women in STEM.

Many things have changed in the past few years in our society. It might not be the case anymore that women are experiencing unfairness in the workplace or school, so we would like to do more research on the most recent state of our admission rate. In addition, we would like to look at the number of admission rates for people with different ethnicities and backgrounds. By knowing the answers to these questions, we can inform and educate people about fairness in today’s education and work’s system.

# Data Cleaning

Throughout this part, we focus on:
- Loading the five datasets
- Checking for missing values
- Extracting only the desired information that constitutes of STEM careers for both men and women
- Creating dataframes that reflect the difference in both the number of Master Degree Awarded and Employment between men and women

```python
# Load the first dataset about the number of men and women in STEM majors
df_STEMGRAD = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/women-stem.csv')
df_STEMGRAD

# Checking if there is any missing value
df_STEMGRAD.isnull().sum()
```

```python
# Drop unnecessary columns, only keep Major Category, Men and Women columns
df_STEMGRAD = df_STEMGRAD.drop(columns = ['Rank', 'Major_code', 'Major', 'Total', 'ShareWomen', 'Median'], inplace=False)
df_STEMGRAD
```

```python
# Sum all numbers of men and women in each major category
df_STEMGRAD_TOTAL = df_STEMGRAD.groupby(['Major_category'])['Men','Women'].sum()
df_STEMGRAD_TOTAL
```

```python
# Load the second dataset of the number of Master Degree awarded to both men and women from 2006 to 2016
df_MASTERGRAD = pd.read_excel('dataset2.xlsx', header = 3)

#Convert to csv
df_MASTERGRAD.to_csv(r'dataset2.csv', index = None, header=True)
df_MASTERGRAD = pd.read_csv('dataset2.csv')

# Locate the row that combines all STEM majors (S&E)
df_SE_TOTAL = df_MASTERGRAD.iloc[1]
df_SE_TOTAL
```

```python
# Check if there is any missing value
df_MASTERGRAD.isnull().sum()
```

```python
# Load the third dataset of number of Master Degree awarded to only women from 2006 to 2016
df_WOMEN = pd.read_excel('dataset3-women.xlsx',header = 3)
df_WOMEN = df_WOMEN.drop(['% female', 'Unnamed: 13'], axis = 1)
df_WOMEN = df_WOMEN.drop(0,axis = 0)

#Convert to csv
df_WOMEN.to_csv(r'dataset3-women.csv', index = None, header=True)
df_WOMEN = pd.read_csv('dataset3-women.csv')
df_WOMEN

# Locate only the row that has the numbers of women receiving Master Degree from STEM majors
df_SE_WOMEN = df_WOMEN.iloc[1]
df_SE_WOMEN
```

We decided to only use the years 2008 to 2016 because in the Master Degree Awarded datasets,
the years range from 2006 to 2016. However, in the STEM career employment datasets, the years 
range from 2008 to 2017. Hence we pick our range to have the minimum year of 2008 and the 
maximum year of 2016 in order to utilize all the datasets.

```python
# Create a dataframe that shows number of men vs women receiving Master Degree
list1 = df_SE_WOMEN[3:].tolist()
list2 = df_SE_TOTAL[3:].tolist()
list3 = list()
for i in range(0,len(list1)):
    ''' Getting the number of Master Degree awarded to men by subtracting the number of Master 
    Degree awarded to women from the total number of Master Degree awarded to both men and women'''
    list3.append(list2[i]-list1[i])
    
data = {'Men':list3,'Women': list1, 'Total': list2}
df_SE = pd.DataFrame(data = data)
df_SE.insert(0, 'Year', range(2008, 2008 + len(df_SE)))
df_SE['Ratio']=df_SE['Women']/df_SE['Men']
df_SE
```

```python
# Load the fourth dataset of number of employed women in a variety of occupations from 2008-2016
df_WomenEmploy = pd.read_excel('employed-women.xlsx', header = 3)
df_WomenEmploy = df_WomenEmploy.drop(['% women', 'Unnamed: 12'], axis = 1)
df_WomenEmploy = df_WomenEmploy.drop(0,axis = 0)

#Convert to csv
df_WomenEmploy.to_csv(r'employed-women.csv', index = None, header=True)
df_WomenEmploy= pd.read_csv('employed-women.csv')

# Drop rows that do not constitute of STEM careers 
rowToDrop = [0,1,2,3,7,12,20,21,26,27]
df_WomenEmploy.drop(rowToDrop, axis = 0, inplace = True)
df_WomenEmploy.reset_index(drop = True)

# Only include the years from 2008-2016  
df_WomenEmploy.drop(columns = ['2017'], inplace = True)
df_WomenEmploy
```

```python
# Sum all of the numbers of employed women in STEM for each year from 2008-2016
list_Women = df_WomenEmploy.sum().tolist()
list_Women = list_Women[1:]
list_Women
```

```python
# Load the fifth dataset of the number of employed men in a variety of occupations from 2008-2016
df_MenEmploy = pd.read_excel('employ-men.xlsx', header = 3)

#Convert to csv
df_MenEmploy.to_csv(r'employ-men.csv', index = None, header=True)
df_MenEmploy= pd.read_csv('employ-men.csv')

# Drop rows that do not constitute of STEM careers 
rowToDrop = [0,1,2,3,7,12,20,21,26,27]
df_MenEmploy.drop(rowToDrop, axis = 0, inplace = True)
df_MenEmploy.reset_index(drop = True)

# Only include the years from 2008-2016  
df_MenEmploy.drop(columns = ['2017'], inplace = True)
df_MenEmploy
```

```python
# Sum all of the numbers of employed men in STEM for each year from 2008-2016
list_Men = df_MenEmploy.sum().tolist()
list_Men = list_Men[1:]

# Create a dataframe that shows the number of men vs women employed in STEM from 2008-2016
data2 = {'Men':list_Men, 'Women': list_Women}
df_2 = pd.DataFrame(data = data2)
df_2.insert(0,'Year', range(2008,2008+len(list_Men)))
df_2['Total']=df_2['Women']+df_2['Men']
df_2['Ratio']=df_2['Women']/df_2['Men']
df_2
```

## Linear Regression

In this section, we seek to visualize and compare two models that compare the trend of the ratio of Women vs. Men in Master Degree awarded in STEM majors and the trend of the ratio of Women vs. Men employed in STEM careers.

```python
ax_ratio_1 = df_SE.plot.scatter(x='Year', y='Ratio')
ax_ratio_1.set_title(' Ratio of Women vs. Men in Master Degree Graduation in STEM')
x = df_SE['Year'].to_numpy()
y = df_SE['Ratio'].to_numpy()
p1, p0 = np.polyfit(x, y,deg=1)
linear_x = np.array([np.min(x), np.max(x)])
linear_y = p1 * linear_x + p0
ax_ratio_1.plot(linear_x, linear_y)
```

```python
outcome, predictors = patsy.dmatrices('Ratio ~ Year', df_SE)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

```python
ax_ratio_2 = df_2.plot.scatter(x='Year', y='Ratio')
ax_ratio_2.set_title(' Ratio of Women vs. Men in Employment in STEM')
x = df_2['Year'].to_numpy()
y = df_2['Ratio'].to_numpy()
p1, p0 = np.polyfit(x, y,deg=1)
linear_x = np.array([np.min(x), np.max(x)])
linear_y = p1 * linear_x + p0
ax_ratio_2.plot(linear_x, linear_y)
```

```python
outcome, predictors = patsy.dmatrices('Ratio ~ Year', df_2)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

# Data Analysis & Results

From the first data visualization indicating college majors’ categories versus gender (Visualization #1), we can observe that there are much more women in the process of obtaining degrees in Biology and Life Sciences as well as Health. In contrast, there are much more men in the process of obtaining degrees in Computers, Mathematics, and Engineering compared to women while in the Physical Sciences category, the amount of men and women is about the same. This indicates an interesting insight into women’s preference among STEM majors which seems to be those that involve some form of study of biology. On the other hand, men’s preference among STEM majors tends to lean towards ones that involve some form of mathematics and engineering. 

From the second data visualization indicating the number of Master Degree recipients in STEM majors per gender (Visualization #2), we can observe that the total number of Master degrees awarded for both genders increase each year from 2008 to 2016. However, the scatter plot of the ratio of Master Degree awarded to both women versus men shows a decreasing trend, indicating a growing gap between the number of women and men who receive Master degrees in STEM majors despite the total amount being awarded is constantly rising.

Even though there is a growing gap between the number of women and men receiving Master degrees in STEM, the bar chart showing the number of employed men versus women in STEM (Visualization #3) seems to indicate that even though men got more employment than women in all of the years except for the year 2015, there is no significant difference between the two amounts, except for the year 2008 in which significantly more men got employed than women in STEM. Furthermore, the scatter plot representing the ratio of women versus men in employment in STEM shows an increasing trend in the equality of women and men employment. In fact, in the 9 years that this data spans, the ratio of women employment in STEM compared to men is above 0.98 in 7 of the years, which is within a reasonable margin that we consider the job admission rate of women in STEM to be fair.

For our prediction of a relationship between the ratio of Women versus Men and the predictor of year in the graduation rate and Employment rate, we are going to use the alpha value of 0.01 (alpha = 0.01). 

The first model, we obtain the P-value of 0.332 which is significantly greater than our alpha value so we can use the year predictor to predict the ratio in graduation rate. The slope of the linear regression is negative so we can predict that although the number of women graduating in STEM fields might be increasing year by year but the increase in the number of men in these fields will be significantly larger.

The second model, we obtain the P-value of 0.07 which is also relatively greater than our initial alpha value. Therefore, we can also use this model to predict the ratio in employment rate based on year. Although the first model led us to the conclusion that there will be more men coming to the STEM fields than women but the number of women getting jobs in these fields is increasing and the number of women is almost the same as the number of men in recent years. 

# Ethics & Privacy

With our conclusion that the job acceptance rate for women in STEM fields is not too different from that of men’s, an ethical concern can be raised. The potential result could be erroneously used as a piece of evidence by interest groups that aim to reduce or scrap funding to programs that assist women in choosing a career in STEM fields. 

Bias: There is a bias in our topic as we chose the United States as one of the countries to do analysis on and not including other countries. Secondly, we narrow down our research to be in STEM majors only. However, we aim to find the answer to our hypothesis with actual data and analysis and by no means we want to answer our question with personal biases towards gender or ethnicity. By knowing the answer to our question, this can inform people about the fairness of our current admission rate.

Personal privacy: the data we used are open sources and are cited accordingly. The data relating to job acceptance rates are collected by reputable sources that have been trusted to respect the privacy of the companies that provided the data.

Potential misuse: If not done correctly, people can use our data to reinterpret the results in a negative light towards their bias gender or ethnicity. Our goal is to find the answer to the question of whether admission rates in the workplace and schools are fair for everybody from different ethnicity, gender, and background. We have tried to be as clear and objective as possible in order to prevent potential data misuse.

In the beginning of this project, we asked the question “Are the recent admission rates of women in STEM careers in the United States fair?” Our hypothesis states that "the current job admission rate in STEM fields for women do not adequately correspond to the potential that women graduated with STEM degrees can bring to the workplace."
