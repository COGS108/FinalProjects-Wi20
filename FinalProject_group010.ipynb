# Research Question

What type of post-secondary United States educational programs provide the highest return on investment (ROI) for students receiving Title IV Financial Aid? Additionally which maximizes the ROI for those studying higher-earning potential majors (STEM and Business) and lower-earning potential majors (Social Science, Humanities, Agriculture/Natural Resources, Communications)? 

## Background and Prior Work

Choosing a finacially optimal postsecondary education program, especially in a specific area of study, has been a question circling students and parents for decades. This becomes even more imperative for students who would be receiving financial aid (with/without interest). We wanted to direct our study to students receiving financial aid, by using a very recent dataset released by the U.S. Department of Education in its annual College Scorecard (Nov. 2019). Specifically the students in our dataset all have received Title IV Federal Financial Aid at their institution and program. Title IV specifically includes Direct Subsidized/Unsubsidized Loans, Direct Graduate Loans PLUS loans, Federal Pell Grant, Federal Supplemental Educational Opportunity Grant, and Federal Perkins Loan, all which require a return except the grants allocated to students from low income brackets.  
    
   Previous similar studies have been conducted. One of the very recent studies by the *Business Insider* demonstrated through analysis of the Bureau of Labor Statistics (BLS) data, including a calculated (projected) 40-year income and return on investment (ROI) percentage, that an associate's degree had the highest ROI with 8,459% (**Rosenberg**). ROI here was calculated by dividing the 40-year projected income by total cost of attendance. This study, however, only considered education level and not the factor of varying earning potentials, which has shown dependence on major selection. 
   
   In another study by the *The Balance Careers* site, BLS data was also summarized including average income by degree level with a clear positive correlation between education level and income as. However, *The Balance Careers* site also links a study of survey data from the National Association of Colleges and Employers from 2019 which demonstrated that the higher earning, higher demand brackets for average income were Engineering at 66521 USD, Computer Science at 66005 USD, Math and Science at 61867 USD, Physics at 69900 USD, Marketing at 62624 USD while lower earning and not highly in demand (not indicated in most top 10 most wanted positions/majors) majors included Humanities at 56688 USD, Social Science at 56720 USD, and Communications at 51448 USD. Business at 56720 USD and Agriculture/Natural Resources at 53565 USD had the greatest year over year increase with 4.1 percent and in a high demand respectively, so we will consider both of these as high earning potential majors (**Doyle**). It is important to remember, as mentioned in the study, that there are clearly several other factors which contribute to earning potential such as skills, personality, values, interests, etc and that major choice is obviously not the only thing that affects earning potential. But since, our study is based on optimizing education level based on field of study, considering major is very critical here. 
  
   The education levels in the data set which are considering are Associates, which are two-year programs including community colleges, technical colleges, and vocational(trade) schools, Bachelors which are four-year university or college programs, Masters which are typically one or two year programs requiring a Bachelor's or Associate's Degree, and Doctorate which, for our study, includes Doctorate of Philosophy (PhD). We are not including Professional Doctorates such as Doctor of Pharmacy, Medicine, Law, etc. due to a high variability in cost of attendance, length of attendance required, and earning potential which would unnecessarily complicate our question and should probably be left alone as its own study. Additionally we are not considering online degree programs or professional certificates for similar reasons in cost of attendance variability and are not traditional length programs. 
   
   For our question we define ROI to be a projected income (40-year) over the debt mean for attendance. As noted in our datasets documentation, if a student attended postsecondary education previously this is also included as a cumulative debt measurement for that observation. To distinguish between high-earning and low-earning majors, we will initially assume the distinguishment discussed in the *The Balance Careers* 2019 study but will confirm this using visualization of salaries. We will then extend our analysis to maximizing ROI for both education and differences in earning potential by major.

# Hypothesis


We hypothesize that for high-earning potential majors (STEM and Business), that a Bachelor's degree will provide a maximized ROI due to a seemingly strong balance between increased higher-earning opportunities relative to an Associate's and lower relative debt for a Master's or non-funded Doctoral Degree. For lower-earning potential majors, we predict that an Associate's Degree (community, technical, vocational, 2-year college/university programs) will provide a minimized opportunity cost due to lowest risk in terms of the tradeoff between debt and a lower relative earning potential/demand field.

### Create a standardize function to assign boolean values to a new column that determines if the major is STEM/business.

```python
def standardize_stem_or_bus(s):
    s = s.lower()
    s = s.strip()
    
    if 'liberal arts' in s:
        return False
    elif 'nursing' in s:
        return True
    elif 'physics' in s:
        return True
    elif 'health' in s:
        return True
    elif 'physical' in s:
        return True
    elif 'sales' in s:
        return True
    elif 'manage' in s:
        return True
    elif 'research' in s:
        return True
    elif 'agri' in s:
        return True
    elif 'veterinary' in s:
        return True
    elif 'science' in s:
        return True
    elif 'molecular' in s:
        return True
    elif 'information operations' in s:
        return True
    elif 'entreprenur' in s:
        return True
    elif 'chem' in s:
        return True
    elif 'computer' in s:
        return True
    elif 'psych' in s:
        return True
    elif 'electrical' in s:
        return True
    elif 'electronic' in s:
        return True
    elif 'data processing' in s:
        return True
    elif 'physics' in s:
        return True
    elif 'bio' in s:
        return True
    elif 'business' in s:
        return True
    elif 'engineer' in s:
        return True
    elif 'business' in s:
        return True
    elif 'medical systems' in s:
        return True
    elif 'math' in s:
        return True
    elif 'technolo' in s:
        return True    
    elif 'marketing' in s:
        return True
    else:
        return False
    
    
```

### Create a function to calculate the ROI for each row

```python
def calculate_roi(x):
    return int((40 * int(x['MD_EARN_WNE'])) / int(x['DEBTMEAN']))
```

### Apply the function by taking the inputs from the "CIPDESC" column and assign the boolean values to a new column called "STEM_OR_BUSINESS"

```python
df_1.assign(STEM_OR_BUSINESS=False)
df_1['STEM_OR_BUSINESS'] = df_1['CIPDESC'].apply(standardize_stem_or_bus)
```

### Apply the calculate_roi function to each row

```python
df_1.assign(ROI=0)
df_1['ROI'] = df_1.apply(calculate_roi, axis=1)
```

# Ethics & Privacy

A common assumption about job earning potential is the choice of majors. As shown in the The Balance Careers in the Prior Work Section, people with high-earning potential majors (STEM and Business) have relatively higher salaries. However, people tend to neglect the impact of education level. Therefore, we have decided to analyze the effect of different degree levels (from Associates to Doctorate) on earning potential for different majors (STEM and Business & non STEM and Business).

Our data is reliable. Even though it is sensitive that our samples are Title IV Federal Financial Aid holders, we do not need to worry about informed consent because we got the data from the U.S. Department of Education College Scorecard. It is a public government official dataset, so we have permission to use the data. Since we only analyze the relationship between education levels and earning potential, the data is anonymous and there are no privacy concerns. We also have 216,638 lines of observations before data cleaning, therefore we have a wide scope of data to make reliable inferences. Last, our sample population has postsecondary education, which is pretty common and spreads all out evenly. So there is no particular population or hidden biases in terms of population demographics.

One unintended consequence is that the audience may assume a college major is the only factor of earning potential and neglect other essential factors such as working experience. Readers may also think this analysis is not comprehensive for not including all education levels, such as Professional Doctorates. To avoid the misunderstanding, we detailedly explain the reasons in the Background and Prior Work Section at the very beginning of the research.

Another concern is that the result may affect some students’ college decisions. They may choose the education with the highest ROI and stop pursuing a better one when they have the opportunity to do so. But since the students in our dataset receive Title IV Federal Financial Aid, our intended audience is actually the students who are eligible for financial aid. Therefore, identifying which education level has the highest return helps this specific population to optimize the limited resources and this research does not perfectly apply to other students who are socioeconomically independent.

Since our analysis is based on both prior work and an authoritative dataset, the research result is convincing and trustworthy.

    Associates Degrees provides the highest return on investment without consideration of Field of Study.

#### Discussion of results   
   For our ROI calculation we assumed a uniform 40-year earning based on the post-graduation 1-year salary ( we never considered salary increases, some fields which may have higher increases than others). We did this for consistency in our analysis.
   
   Limitation:
   We could improve our regression model if we considered salary increase percentages among Fields (finer detail), but from the large dataset and our hypothesis, making an assumption of a constant 40-year earning made our analysis much simpler.
   
   From all the data and visualization we see that the opportunity cost is relatively higher for higher degrees (Doctoral vs Associates).  We don't conclude which degree or major is better at salary, but from return on investment instead. Even though the higher the degree the higher the earnings, they still had to invest more. Compared to what they invested, we see that for both STEMB and non STEMB majors, Associate Degree got the highest return on investment. 
   
   Our result for STEMB majors contradicts our hypothesis, we hypothesized that Bachelor's degree would have the highest ROI because it has higher earnings and not as much debt compared to higher degrees. Compared to the earnings, the opportunity cost is really low. But probably because we assumed a uniform 40-year earning based on the post-graduation 1-year salary, our results do not match our hypothesis.
