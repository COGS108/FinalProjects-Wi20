1) Preventative Services: Annual Wellness Checkup

2) Cases of Reported Hospitalization

3) Number of Inpatient Days

We finally developed linear models to draw conclusions about the correlation between observed discrepancies and socioeconomic status to determine the validity of our overarching hypothesis.

# Research Question

The research question for this project can be summarized as follows:
#### *Are there disparities (measured based on the parameters: preventative services, hospitalization, inpatient days, and mortality rate) in the quality of medical care in the United States when looking at socioeconomic status, state, and/or ethnicity.”*

Note: We use an alpha value of 0.05  during hypothesis testing. This gives a 5% chance that the null hypothesis is actually true in the case that it is rejected. Though we will not further examine these conclusions, it is interesting to note that even twenty years ago, efforts were being made to minimize the discrepancies seen in medicine even at that time. # Data Cleaning

After setting up, we realized that a majority of our data was seperated into a very large amount dataframes, which would be hard to manage as we continued to work on the project. We chose to combine these datasets to better clarify what data was useable and relavent.

Below is the code that we used to clean the datasets, and eventually merge them to create 3 major dataframes.
We create 3 major dataframes: `df_states`, `Ethnic_income`,  `race_percent`

Note that `income_all` initially included data from more than just the year 2017, we removed all data except for 2017 as the rest of our data is from the year 2017 as well.

Additionally, we also create dataframes for each parameter (hospitalization, inpatient days, and preventative services)
with respect to each race (White, Black, Asian, Hispanic, and Native American), which we used to further our analysis.

```python
## Cleaning 'income_all'
#Remove all instances of 'Total' from the income_all Dataframe
Base = income_all[(income_all.Race != "Total")][["Race", "State", "Year", "Household Income by Race"]]

Base
```

```python
## Cleaning & Merging 'Ethnic_income'
#Isolate dataframe with the income of White people
White_income = Base[Base.Race == "White"][["Race", "State", "Year", "Household Income by Race"]].drop(columns = ["Race"]).rename(columns={"Household Income by Race": "White Income"})
#Isolate dataframe with the income of Black people
Black_income = Base[Base.Race == "Black"][["Race", "State", "Year", "Household Income by Race"]].drop(columns = ["Race"]).rename(columns={"Household Income by Race": "Black Income"})
#Isolate dataframe with the income of Asian people 
Asian_income = Base[Base.Race == "Asian"][["Race", "State", "Year", "Household Income by Race"]].drop(columns = ["Race"]).rename(columns={"Household Income by Race": "Asian Income"})
#Isolate dataframe with the income of Native American people 
Native_income = Base[Base.Race == "Native American"][["Race", "State", "Year", "Household Income by Race"]].drop(columns = ["Race"]).rename(columns={"Household Income by Race": "Native Income"})
#Isolate dataframe with the income of Hispanic people 
Hispanic_income = Base[Base.Race == "Hispanic"][["Race", "State", "Year", "Household Income by Race"]].drop(columns = ["Race"]).rename(columns={"Household Income by Race": "Hispanic Income"})
#Merge all income dataframes into one dataframe
Ethnic_income = Hispanic_income.merge(Native_income).merge(Asian_income).merge(Black_income).merge(White_income)
#Removing All years except for 2017
Ethnic_income = Ethnic_income[Ethnic_income.Year == 2017].reset_index().drop(columns = ["index"])
#Sort Values based on 'State'
Ethnic_income = Ethnic_income.sort_values ('State')
#Reset the Index
Ethnic_income = Ethnic_income.reset_index(drop = True)

Ethnic_income.head()
```

```python
## Cleaning 'race_percent'

#Change entries from Decimal to Percent 
for x in race_percent.columns[1::]:
    race_percent[x] = race_percent[x] * 100
    
race_percent.head()
```
