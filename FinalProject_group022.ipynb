```python
#Creates an empty Dataframe to hold revenues for the top rangeTop amount of companies each year
def createNewDF(rangeTop):
    column_values = ['year']

    #Each ranking for each year has 5 peices of information (columns) associated
    for i in range(1, rangeTop + 1):
        column_values.append('#' + str(i) + ' Revenue')
        column_values.append('#' + str(i) + ' Company')
        column_values.append('#' + str(i) + ' Matched Company')
        column_values.append('#' + str(i) + ' Sector')
        column_values.append('#' + str(i) + ' State')
    
    df_raw_revenue = pd.DataFrame(columns=column_values)
    #set index to be year
    df_raw_revenue.set_index('year', inplace=True)
    return df_raw_revenue
```

```python
#Given name of the company, extracts its corresponding best match,sector, and state using the SP_Data
def getCompanyInfoUsingCSV(company, df_company_info, yearStr, rank):
    dict_states = getDictStates()
    
    #attempts to get the best match for company, if it exists within the CSV file
    try:
        #To match, 70% of the string must be the same
        best_match = difflib.get_close_matches(company, df_company_info['Security'], cutoff=0.7)[0]
    #No such match exists, returns None for all
    except:
        return None, None, None
    #gets index in the csv file for the best company match
    index = df_company_info[df_company_info['Security'] == best_match].index.tolist()[0]
    #gets the headquarter information, extracting the state information into a consistent format
    state_abr = dict_states[df_company_info.loc[index]['Headquarters Location'].split(',')[-1].strip()]
    return best_match, df_company_info.loc[index]['GICS Sector'], state_abr

#Checks if given sector is considered to be under the umbrella of "Technology", returns True if yes
def checkForTechnologySector(sector):
    if sector == None:
        return None
    #Check sector Technology
    if sector in ['Information Technology', 'Technology', 'Telecommunications']:
        return True
    return False
```

```python
#Accounts for inflation of given revenue in a given year (standarizes to Novemeber 2019)
def inflateRevenue(revenue, year):
        return cpi.inflate(revenue, year)
    
#Populates empty dataframe with the information from the Fortune500 csv files, filling in company's name,
#its best match from SP_Data, its sector (SP_Data), its state symbol (SP_Data),and its inflated revenue (millions)
def populateRawDF(df, startYear, endYear, rangeTop):
    dict_sectors = {}
    company_info = pd.read_csv('fortune500/csv/SP500_Data/S&P500-Condensed.csv')
    #loops through every desired year from Fortune 500
    for year in range(startYear,endYear + 1):

        #reads in corresponding Fortune500 csv file for that year
        temp = pd.read_csv('fortune500/csv/fortune500_csv/fortune500-'+ str(year) + '.csv')
        #extrats the top desired amount of rows
        temp = temp.head(rangeTop)
        
        #gets revenue information for top desired amount of rows
        revenues_list = list(temp['revenue ($ millions)'])
        
        #traverses individually through top desired amount of rows in that year
        for index in range(rangeTop):
            
            #extracts compnay information for that ranking
            company = temp.loc[index]['company']
            #gets its best matched company (and thereby sector & state information) from SP_Data
            matched_company, sector, state_abr = getCompanyInfoUsingCSV(company, company_info, str(year), str(index + 1))
            #determines if the company is a "Technology" company
            is_information_sector = checkForTechnologySector(sector)
            
            #accounts for inflation for the revenue 
            inflated_revenue = inflateRevenue(revenues_list[index], year)
                  
            #Sets the corresponding column with company's revenue information
            df.set_value(year, '#' + str(index + 1) + ' Revenue', inflated_revenue)
            
            #Sets the corresponding column with company's name
            df.set_value(year, '#' + str(index + 1) + ' Company', company)
            
            #Sets the corresponding column with company's best matched compay 
            df.set_value(year, '#' + str(index + 1) + ' Matched Company', matched_company)
            
            #Sets the corresponding column with company's sector information
            df.set_value(year, '#' + str(index + 1) + ' Sector', sector)
            
            #Sets the corresponding column with company's state symbol information
            df.set_value(year, '#' + str(index + 1) + ' State', state_abr)
```

```python
#returns a new empty dictionary containing state symbols as keys, and 0 as their values
#Used to help keep count of companies while parsing infromation in constructStateIndexDataFrame method
def getNewStatesDict():
    dict_states = {}
    for state in getStates():
        dict_states[state.strip()] = 0
    return dict_states

#Creates an empty dataframe to hold the final dataframe values, namely the average revenue (values) for a certain
#amount of years (columns) indexed by states (rows)
def initializeStateIndexDataFrame(startYear, endYear):
    column_values = ['state']

    for year in range(startYear, endYear + 1):
        column_values.append(str(year))
    
    df = pd.DataFrame(columns=column_values)
    df.set_index('state', inplace=True)
    return df
    
#Creates final dataframe with the average revenue for a particular state and year 
def constructStateIndexDataFrame(df, startYear, endYear, rangeTop, is_Tech_Df):
    #creates an empty dataframe with the rows and columns pre-filled
    df_final = initializeStateIndexDataFrame(startYear, endYear)
    
    #Goes through each year desired year (column)
    for year in range(startYear, endYear + 1):
        
        #initializes two new dictionaries for each year
        dict_rev = getNewStatesDict() #tracks total revenue for that state
        dict_count = getNewStatesDict() #tracks total companies in that state
        
        #Goes through each rank in that year 
        for rank in range(1, rangeTop + 1):
            
            #gets the index of that rank in the passed in dataframe
            index = year - startYear;
            #Extracts the revenue information for that company from passed in dataframe
            revenue = float(df.loc[index]['#' + str(rank) + ' Revenue'])
            #Extracts the sector information for that company from passed in dataframe
            sector = str(df.loc[index]['#' + str(rank) + ' Sector']).strip()
            #Extracts the state information for that company from passed in dataframe
            state = str(df.loc[index]['#' + str(rank) + ' State']).strip()
            
            #checks if sector matches up with inputted parameter
            if checkForTechnologySector(sector) == is_Tech_Df:
                #update revenue information for that state
                dict_rev[state] += revenue
                #update company count for that state
                dict_count[state] += 1
        
        #traverse through filled in dictionary for current year
        for state in dict_rev:
            mean = None
            #check if state had any companies matched to it
            if dict_count[state] > 0: 
                #update mean based off of revenue and count information
                mean = dict_rev[state]/dict_count[state]
            
            #set final mean in dataframe for that year and state
            df_final.set_value(state, str(year), mean)
            
    return df_final
```

## Data Analysis and Predictive Model

__Since we are now considering three variables, region, year, and revenue (log transformed) to predict CO2 emissions (log transformed), we will use patsy.dmatrices to perform a multivariate regression and make a linear model.__

```python
outcome, predictors = patsy.dmatrices('CO2emissions10 ~ revenue10 + region + years', df_revenue_co2)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
```

```python
print(res.summary())
```

## Results

After observing the above OLS regression results, we can see that the R-squared is 0.560 which means that 56% of the variability in CO2 emissions can be explained by the revenue, region and years. This is relatively high.

Based on the results above, however, non-tech companies' revenues do not significantly predict CO2 emissions because the p-value is greater than 0.05 (for our hypothesis, we said a p-value less than 0.05 is statistically significant). However, time(years) does significantly predict CO2 emissions because its p-value is less than 0.05.

From this we can conclude that the revenue of Non-tech companies does not predict CO2 emissions in the U.S. This finding shows our hypothesis that revenue influences CO2 emissions to be false.
